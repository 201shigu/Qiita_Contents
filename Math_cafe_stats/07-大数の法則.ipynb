```python
def create_average_ts(result_list):
    ave_result = []
    ave_val = 0
    for i, res in enumerate(result_list):
        n = i+1
        ave_val = (ave_val*(n-1)+res)/n
        ave_result.append(ave_val)
    return ave_result
```

```python
rd.seed = 71
n_sample = 1000000
p =0.5
plt.figure(figsize=(14,6))
for i in range(20):
    x = st.bernoulli.rvs(p, size=n_sample)
    print(np.average(x), end=", ")
    x_ave = create_average_ts(x)
    plt.plot(x_ave, alpha=0.7)
    plt.xscale("log")
    print(x_ave[-1])
```
